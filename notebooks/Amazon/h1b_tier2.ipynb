Top Job Titles that Have Increased the Most between 2020-2025

```python
#restrict to jobs that show up >1k times in 2021, or else you get weird, misc titles
def standardize_title(series):
    return (
        series.dropna()
        .astype(str)
        .str.strip()
        .str.replace(r'\s+', ' ', regex=True)
        .str.upper()
    )

counts_2021 = standardize_title(df_2021['JOB_TITLE']).value_counts().rename('count_2021')
counts_2025 = standardize_title(df_2025['JOB_TITLE']).value_counts().rename('count_2025')

comparison = pd.concat([counts_2021, counts_2025], axis=1).fillna(0)
comparison[['count_2021', 'count_2025']] = comparison[['count_2021', 'count_2025']].astype(int)

eligible = comparison[comparison['count_2021'] >= 1000].copy()

increased = eligible[eligible['count_2025'] > eligible['count_2021']].copy()
increased['pct_increase'] = ((increased['count_2025'] - increased['count_2021']) / increased['count_2021']) * 100

top_10_increases = increased.sort_values('pct_increase', ascending=False).head(10)

result = (
    top_10_increases[['count_2021', 'count_2025', 'pct_increase']]
    .reset_index(names='JOB_TITLE_STD')
    .assign(pct_increase=lambda x: x['pct_increase'].round(2))
)

result
```

```python
top_n = 20  

# Restrict to AI/computer systems realm (SOC major group 15-xxxx)
soc_col_candidates = ['SOC_CODE', 'SOC']
soc_col = next((c for c in soc_col_candidates if c in df_2021.columns and c in df_2025.columns), None)

mask_2021 = df_2021[soc_col].astype(str).str.strip().str.startswith('15-')
mask_2025 = df_2025[soc_col].astype(str).str.strip().str.startswith('15-')

df_2021_realm = df_2021[mask_2021].copy()
df_2025_realm = df_2025[mask_2025].copy()

titles_2021 = set(standardize_title(df_2021_realm['JOB_TITLE']))
titles_2025 = standardize_title(df_2025_realm['JOB_TITLE'])

new_in_2025_counts = (
    titles_2025[~titles_2025.isin(titles_2021)]
    .value_counts()
    .head(top_n)
)

new_in_2025_result = (
    new_in_2025_counts
    .rename_axis('JOB_TITLE_STD')
    .reset_index(name='count_2025')
)

new_in_2025_result
```

```python
plot_counts = new_in_2025_counts.head(20).sort_values()

plt.figure(figsize=(12, 8))
plot_counts.plot(kind='barh')
plt.title('Top 20 Job Titles in 2025 Not Present in 2021')
plt.xlabel('Count in 2025')
plt.ylabel('Standardized Job Title')
plt.tight_layout()
plt.show()
```
